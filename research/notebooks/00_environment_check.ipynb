## サイン波の生成と描画

440 Hz（ラ音）のサイン波を生成して波形を表示する。

```python
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 44100
duration    = 0.02   # 20 ms
freq        = 440.0  # Hz

t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
wave = np.sin(2 * np.pi * freq * t)

plt.figure(figsize=(10, 3))
plt.plot(t * 1000, wave)
plt.xlabel("Time [ms]")
plt.ylabel("Amplitude")
plt.title(f"Sine wave ({freq} Hz)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## FFT によるスペクトル確認

```python
from scipy.fft import rfft, rfftfreq

N = len(wave)
spectrum = np.abs(rfft(wave)) / N
freqs    = rfftfreq(N, 1 / sample_rate)

plt.figure(figsize=(10, 3))
plt.plot(freqs, spectrum)
plt.xlim(0, 1000)
plt.xlabel("Frequency [Hz]")
plt.ylabel("Amplitude")
plt.title("FFT Spectrum")
plt.grid(True)
plt.tight_layout()
plt.show()
```
